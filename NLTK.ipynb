how VADER differs form Affin and TextBlob?

ADER, AFINN, and TextBlob are three widely used tools for sentiment analysis, but they differ in their approach and suitability for different types of text. VADER (Valence Aware Dictionary for Sentiment Reasoning) is a lexicon and rule-based model developed to analyze sentiments in social media content such as tweets, comments, and short messages. It can interpret emojis, capitalization, punctuation, and degree modifiers like “very” or “extremely,” which makes it effective for informal language. VADER provides four sentiment scores—positive, negative, neutral, and compound (ranging from -1 to +1)—and is well-suited for detecting intensity and negations in text.

```python
# Install libraries (run once)
# !pip install nltk textblob afinn

import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from afinn import Afinn
from textblob import TextBlob

# Download VADER lexicon
nltk.download('vader_lexicon')

# Initialize analyzers
sia = SentimentIntensityAnalyzer()
af = Afinn()

# --- One sarcastic sentence ---
sentences = ["Oh fantastic, another bug to fix right before the deadline 🙃"]

# Create DataFrame
df = pd.DataFrame(sentences, columns=['text'])

# --- AFINN Analysis ---
df['afinn_score'] = df['text'].apply(lambda x: af.score(x))

# --- TextBlob Analysis ---
df['textblob_polarity'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['textblob_subjectivity'] = df['text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
df['textblob_sentiment'] = df['textblob_polarity'].apply(
    lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral')
)

# --- VADER Analysis ---
df['vader_scores'] = df['text'].apply(lambda x: sia.polarity_scores(x))
df['vader_compound'] = df['vader_scores'].apply(lambda x: x['compound'])
df['vader_sentiment'] = df['vader_compound'].apply(
    lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral')
)

# --- Display Results ---
df_display = df[['text', 'afinn_score', 'textblob_polarity', 'textblob_subjectivity',
                 'textblob_sentiment', 'vader_compound', 'vader_sentiment']]

display(df_display)
```

```python
### TASK 2

texts=[
    "I LOVE this movie!!! 😍😍",
    "This course is SO boring... 😴 ",
    "I think AI is amazing 🤖💡",
    "Not sure if I like this. 🤷",
    "Great work!!! Keep it up 🔥🔥"
]
import pandas as pd
df= pd.DataFrame(texts, columns=["texts"])
df

df['clean_text'] = df['texts'].str.lower()
df['clean_text'] = df['clean_text'].str.strip()
df[['texts','clean_text']]

import string
df['no_punct'] = df['clean_text'].str.translate(str.maketrans('', '', string.punctuation))
df[['clean_text','no_punct']]

nltk.download('stopwords')
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
df['no_stopwords'] = df['no_punct'].apply(lambda x: " ".join([word for word in x.split() if word not in stop]))
df[['no_punct','no_stopwords']]

import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['no_emoji'] = df['no_stopwords'].apply(remove_emojis)
df[['no_stopwords','no_emoji']]
```
